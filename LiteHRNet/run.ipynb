```python
from mmcv import Config

cfg = Config.fromfile(
    './configs/body_2d_keypoint/topdown_heatmap/coco/td-hm_hrnet-w48_dark-8xb32-210e_coco-256x192.py'
)
cfg.data_root = 'dataset/coco'
cfg.work_dir = 'work_dirs/hrnet_w48_dark_coco_256x192'
cfg.gpu_ids = range(1)
cfg.seed = 0


# set batch size
cfg.data.samples_per_gpu = 16
cfg.data.val_dataloader = dict(samples_per_gpu=16)
cfg.data.test_dataloader = dict(samples_per_gpu=16)

# set dataset configs
cfg.data.train.type = 'TopDownCOCOTinyDataset'
cfg.data.train.ann_file = f'{cfg.data_root}/annotations/person_keypoints_train2017.json'
cfg.data.train.img_prefix = f'{cfg.data_root}/images/train2017'

cfg.data.val.type = 'TopDownCOCOTinyDataset'
cfg.data.val.ann_file = f'{cfg.data_root}/person_keypoints_val2017.json'
cfg.data.val.img_prefix = f'{cfg.data_root}/images/'

cfg.data.test.type = 'TopDownCOCOTinyDataset'
cfg.data.test.ann_file = f'{cfg.data_root}/person_keypoints_val2017.json'
cfg.data.test.img_prefix = f'{cfg.data_root}/images/'
```
